```python
import zipfile
import pandas as pd

# Unzip the file
with zipfile.ZipFile('dataset.zip', 'r') as z:
    z.extractall('.')
    print("Files inside:", z.namelist())

# Load the CSV
df = pd.read_csv('stock_prices_daily.csv')
print(f"\nTotal records: {len(df)}")
print(f"\nColumns: {df.columns.tolist()}")
df.head()

```

```python
# Simple AAPL price chart
aapl = df[df['Ticker'] == 'AAPL'].copy()
aapl['Date'] = pd.to_datetime(aapl['Date'])

plt.figure(figsize=(12,5))
plt.plot(aapl['Date'], aapl['Close'])
plt.title('AAPL Closing Price 2020-2026')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
